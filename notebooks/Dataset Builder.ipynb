```python
import pandas as pd
import numpy as np
import json
import re
```

```python
def load_squad_df(squad_dataset):
    squad_data = pd.read_json(squad_dataset)
    reformed = {"contexts": [], "questions": [], "answer_texts": [], "answer_starts": []}

    for i in range(squad_data.shape[0]):
        for paragraph in squad_data.iloc[i,1]["paragraphs"]:
            for qas in paragraph["qas"]:
                answers = []
                for answer in qas["answers"]:
                    if answer not in answers:
                        answers.append(answer)
                        reformed["contexts"].append(paragraph["context"])
                        reformed["questions"].append(qas["question"])
                        reformed["answer_texts"].append(answer["text"])
                        reformed["answer_starts"].append(answer["answer_start"])

    return pd.DataFrame(reformed)  
```

```python
squad_dev_df = load_squad_df("../corpora/dev-v2.0.json")
squad_dev_df.to_csv("../corpora/squad-dev-v2.0.csv")
squad_dev_df.head()
```

```python
squad_train_df = load_squad_df("../corpora/train-v2.0.json")
squad_train_df.to_csv("../corpora/squad-train-v2.0.csv")
squad_train_df.head()
```
