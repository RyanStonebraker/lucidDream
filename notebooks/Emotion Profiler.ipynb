```python
presence = []
presence_threshold = 10
last_seen = {}
current_scene = set()
curr_book = 1
for i, row in main_potters.iterrows():
    if row["book"] != curr_book:
        curr_book = row["book"]
        current_scene = set()
        last_seen = {}
    last_seen[row["character"]] = row["line"]
    current_scene = set([character for character in current_scene if row["line"] <= last_seen[character] + presence_threshold])
    current_scene.add(row["character"])
    presence.append(list(current_scene))
main_potters["scene"] = presence
```
