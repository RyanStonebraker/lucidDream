```python
harry_potter_df = harry_potter_df[~harry_potter_df.character.isin(non_characters)]
```

```python
harry_potter_df.loc[harry_potter_df.character == "scene", "character"] = "environment"
harry_potter_df.loc[harry_potter_df.character == "tom", "character"] = "tom riddle"
harry_potter_df.loc[harry_potter_df.character == "albus", "character"] = "albus dumbledore"
harry_potter_df.loc[harry_potter_df.character == "dumbledore", "character"] = "albus dumbledore"
```

```python
main_characters = ["harry", "hermione", "ron", "albus dumbledore", "snape", "hagrid", "tom riddle"]
```

```python
main_potters = harry_potter_df[harry_potter_df.character.isin(main_characters)]
```

```python
presence = []
presence_threshold = 10
last_seen = {}
current_scene = set()
curr_book = 1
for i, row in main_potters.iterrows():
    if row["book"] != curr_book:
        curr_book = row["book"]
        current_scene = set()
        last_seen = {}
    last_seen[row["character"]] = row["line"]
    current_scene = set([character for character in current_scene if row["line"] <= last_seen[character] + presence_threshold])
    current_scene.add(row["character"])
    presence.append(list(current_scene))
main_potters["scene"] = presence
```
